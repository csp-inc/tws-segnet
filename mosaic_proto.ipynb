# Below cell makes the file structure

```python
def create_subdirectories_from_filenames():
    '''
    Builds subdirectories given a list of unsorted file names and string indices to search
    
    Parameters
    ----------
    
    
    '''
    # this is a funciton to move all the data to the appropriate image directory and date
    for d in dates:
        os.makedirs('%s/%s'%(wd,d), exist_ok=True)
        date_files = [x for x in imgs if '%s'%(d) in x]   
        for i in img_nums:
            os.makedirs('%s/%s/%s'%(wd,d,i), exist_ok=True)
            img_files = [y for y in date_files if '%s'%(i) in y]   
            for q in qq:
                qq_files = [z for z in img_files if '%s'%(q) in z]
                if (len(qq_files)==0): #sometimes we don't have that quadrat
                    continue
                else:
                    os.makedirs('%s/%s/%s/%s'%(wd,d,i,q), exist_ok=True)        
                    [shutil.copy(o,'%s/%s/%s/%s'%(wd,d,i,q)) for o in qq_files if '%s'%(q) in o]
```

Go into the first date directory and get the rows and columns

```python
#gather by the specific i, d, and q
for d in dates:
    for im in img_nums:
        for q in qq:
            data_dir = '%s/%s/%s/%s'%(wd, d, im, q)
            n=len(data_dir)
            subfiles = glob('%s/*.png'%data_dir)
            if (len(subfiles)==0): #sometimes we don't have that quadrat
                continue
            row_nums = list(set([int(i[-43:-39]) for i in subfiles]))
            col_nums = list(set([int(i[-35:-31]) for i in subfiles]))
            R = max(row_nums) #number of rows
            C = max(col_nums) #number of cols
            out_filename = './m_%s_%s_11_1_%s.png'%(im, q, d)
            #now window movement
            
            # 2 nested for loops
            bigcat = []
            for r in range(R):
                # get a list of names for that particular row
                z = [f for f in subfiles if 'row%04d'%r in f]
                row_list = []
                for c in range(C):
                    img = np.array(Image.open([ig for ig in z if 'col%04d'%(c) in ig][0]))
                    row_list.append(img)
                bigcat.append(np.concatenate(row_list, axis=1))
            out = Image.fromarray(np.concatenate(bigcat, axis=0))
            out.save(out_filename)
```
